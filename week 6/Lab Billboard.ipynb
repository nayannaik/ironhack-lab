## Importing Librarties

```python
# 1. import libraries
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
# 2. find url and store it in a variable
url = "https://www.billboard.com/charts/hot-100/"
```

```python
# 3. download html with a get request 
response = requests.get(url)
```

```python
response.content
```

```python
# getting the response status code
response.status_code
```

```python
# 4.1. parse html (create the 'soup')
soup = BeautifulSoup(response.content, "html.parser")
# 4.2. check that the html code looks like it should
#soup
```

## Getting the Songs name

```python
soup.select(".a-no-trucate.a-font-primary-bold-s")[0]
```

#### Iterating it to get all songs and putting it into list

```python
song_name = []
for i in soup.select(".a-no-trucate.a-font-primary-bold-s"):
    print(i.get_text().strip())
    song_name.append(i.get_text(strip=True))
```

## Getting the Songs artist

```python
soup.select(".a-no-trucate.a-font-primary-s")
```

```python
#### Iterating it to get all artists and putting it into list
```

```python
song_artist = []
for i in soup.select(".a-no-trucate.a-font-primary-s"):
    print(i.get_text().strip())
    song_artist.append(i.get_text(strip=True))
```

```python
#printing list just to check
print(song_name)
```

```python
#### Creating a Data Frame
```

```python
df = pd.DataFrame({'song_name' : song_name,
                                'song_artist' : song_artist })
df
```
