```python
def load_original_data():
    file1 = pd.read_csv('file1.csv')
    file2 = pd.read_csv('file2.csv')
    file3 = pd.read_csv('file 3 clean.csv')
    return pd.concat([file1,file2,file3], axis=0) 
```

```python
ica_df=load_original_data()
ica_df
```

```python
def lower_case_column_names(ica_df):
    ica_df.columns=[i.lower() for i in ica_df.columns]
    return ica_df
lower_case_column_names(ica_df)
```

```python
def drop_columns(ica_df) :
    ica_df.drop(columns =['customer'], inplace = True)
    return ica_df
drop_columns(ica_df)
```

```python
ica_df.reset_index(drop=True, inplace=True)
```

```python
ica_df['number of open complaints'].apply(type)
```

```python
def clean_numberofcomps(x):
    if isinstance(x, str):
        return x[2]
    else:
        return x
```

```python
ica_df["number of open complaints"] = ica_df["number of open complaints"].apply(clean_numberofcomps)
```

```python
ica_df
```

```python
ica_df.gender.unique()
```

```python
def clean_gender(x):
    if x in ['M', 'Male']:
        return 'Male'
    elif x in ['F', 'Femal','female']:
        return 'Female'
    else:
        return 'nan'
```

```python
ica_df['gender'] = ica_df['gender'].apply(clean_gender)
ica_df
```

```python
ica_df.st.unique()
```

```python
ica_df['st'] = ica_df['st'].replace({'WA':'Washington', 'AZ':'Arizona','Cali': 'California'})
```

```python
ica_df['customer lifetime value'].apply(type)
```

```python
def clean_customerlifetitmvalue(x):
    if isinstance(x, str):
        return x.removesuffix('%')
    else:
        return x
```

```python

ica_df['customer lifetime value'] = ica_df['customer lifetime value'].apply(clean_customerlifetitmvalue)
```

```python
ica_df['customer lifetime value'] =  pd.to_numeric(ica_df['customer lifetime value'], errors='coerce')
#ica_df['income'] =  pd.to_numeric(ica_df['income'], errors='coerce')
```

```python
ica_df.drop_duplicates()
```

```python
mean_customer_lifetime_value = np.mean(ica_df['customer lifetime value'])
ica_df['customer lifetime value'] == ica_df['customer lifetime value'].fillna(mean_customer_lifetime_value, inplace = True)
ica_df
```

```python
mean_monthly_premium_auto = np.mean(ica_df['monthly premium auto'])
ica_df['monthly premium auto'] == ica_df['monthly premium auto'].fillna('mean_monthly_premium_auto')
ica_df
```

```python
mean_income = np.mean(ica_df['income'])
ica_df['income'] == ica_df['income'].fillna(mean_income, inplace = True)
ica_df
```

```python
ica_df['number of open complaints'] =  pd.to_numeric(ica_df['number of open complaints'], errors='coerce')
```

```python
mean_number_of_open_complaints= np.mean(ica_df['number of open complaints'])
ica_df['number of open complaints'] == ica_df['number of open complaints'].fillna(mean_number_of_open_complaints, inplace = True)
ica_df
```

```python
mean_total_claim_amount = np.mean(ica_df['total claim amount'])
ica_df['total claim amount'] == ica_df['total claim amount'].fillna(mean_total_claim_amount, inplace = True)
ica_df
```
