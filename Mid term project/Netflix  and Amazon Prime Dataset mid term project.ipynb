## Data Cleaning
### - remove the string in the columns that appear as a the first string of the row but is not a genre 
This step is to allow us to be able to classify the different rows in the column listed_in into various genres

```python
netflix['listed_in'] = netflix['listed_in'].str.split(',').apply(lambda x: [cat.strip() for cat in x if 'International' not in cat])
```

```python
netflix['listed_in'] = netflix['listed_in'].apply(lambda x: x[0] if x else 'Others')
```
